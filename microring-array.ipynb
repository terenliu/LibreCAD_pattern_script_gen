```python
import numpy as np
import sys
from common_shape_generator_v3 import xy_disp,ci_disp,rect_disp,mesa_ring_disp,triangle_disp,arc_disp,circle_poly_generator,rect_centre_disp

cell_dist_x=2480
cell_dist_y=1400# we only generate one row. no need for considering y.

FP_centre_x=[150,625,1150,1800]
FP_centre_y=[705,585,465,345]

FP_width=[20,40,60] # for the top row for short dist, narrow width testing, just set three value as 10
#FP_width=[10,10,10]
FP_length=[250,500,750,1000]
#FP_length=[120,300,750,1000]
FP_bottom_contact_window_dist=5
FP_bottom_contact_window_reflector_dist=8 # dist on light emission edge. which should be larger
FP_bottom_contact_window_width=10

FP_bragg_grating_num=8


FP_bragg_grating_gap=450e-3
FP_bragg_grating_bar=150e-3
FP_bragg_grating_period=FP_bragg_grating_gap+FP_bragg_grating_bar

FP_contactwin_edge_dist=6
FP_contact_drc_dist=5 #distance between top contact and bottom contact

FP_top_pad_width=300
layer_name='SiO2-gratings'# or bottomcontact-win-region/bottomcontact-diff (contact window was based on rectangle strip off mesa and top contact pad region), bottom-metal (similar. using contact_diff strip off the top contact). and top contact

def grating_generator(direction,edge_x, centre_y, grating_bar,grating_period,centre_pillar_width,FP_width,grating_number):
    if direction == 'left2right':
        x_sign=1;
    elif direction=='right2left':
        x_sign=-1;
    for i in range(0,grating_number):
        rect_disp(x_left=edge_x+x_sign*i*(grating_period)+x_sign*grating_bar,
                  x_span=x_sign*(grating_period-grating_bar),
                  y_bottom=centre_y+centre_pillar_width/2,
                  y_span=(FP_width/2)-(centre_pillar_width/2)
                )
        rect_disp(x_left=edge_x+x_sign*i*(grating_period)+x_sign*grating_bar,
                  x_span=x_sign*(grating_period-grating_bar),
                  y_bottom=centre_y-centre_pillar_width/2,
                  y_span=-(FP_width/2)+(centre_pillar_width/2)
                )

for i in range(0,len(FP_width)):
    x0=i*cell_dist_x;
    y0=0;
    for j in range(0,len(FP_length)):
        x1=x0+FP_centre_x[j]
        y1=y0+FP_centre_y[j]
        top_contact_width=min(FP_width[i]-8,0.3*FP_width[i])
        mesa_top_y=y1+FP_width[i]*0.5
        mesa_bottom_y=y1+FP_width[i]*0.5
        topcontact_top_y=y1+top_contact_width*0.5
        bottomcontact_bottom_y=y1-FP_width[i]-FP_bottom_contact_window_dist-FP_bottom_contact_window_width


        if layer_name =='mesa':
            rect_centre_disp(xc=x1,yc=y1,x_half_span=FP_length[j]*0.5,y_half_span=FP_width[i]*0.5)
        
        elif layer_name =='grating-gaps':
            grating_generator(direction='left2right',edge_x=x1-FP_length[j]/2,
                              centre_y=y1,grating_bar=FP_bragg_grating_bar,grating_period=FP_bragg_grating_period,
                              centre_pillar_width=top_contact_width+2,FP_width=FP_width[i],
                              grating_number=FP_bragg_grating_num)  
            grating_generator(direction='right2left',edge_x=x1+FP_length[j]/2,
                              centre_y=y1,grating_bar=FP_bragg_grating_bar,grating_period=FP_bragg_grating_period,
                              centre_pillar_width=top_contact_width+1,FP_width=FP_width[i],
                              grating_number=FP_bragg_grating_num)   
            
        elif layer_name =='SiO2-gratings':
            SiO2_Air_grating_air_gap_width=600e-3
            SiO2_Air_grating_period=600e-3+450e-3
            SiO2_Air_grating_num=40
            grating_generator(direction='right2left',edge_x=x1-FP_length[j]/2-25,
                              centre_y=y1,grating_bar=SiO2_Air_grating_air_gap_width,grating_period=SiO2_Air_grating_period,
                              centre_pillar_width=-2,FP_width=FP_width[i]+5,
                              grating_number=SiO2_Air_grating_num)  
            grating_generator(direction='left2right',edge_x=x1+FP_length[j]/2+25,
                              centre_y=y1,grating_bar=SiO2_Air_grating_air_gap_width,grating_period=SiO2_Air_grating_period,
                              centre_pillar_width=-2,FP_width=FP_width[i]+5,
                              grating_number=SiO2_Air_grating_num) 
        elif layer_name =='ebeam-grating-region':
            margin_distance=3
            rect_disp(x_left=x1-FP_length[j]/2-margin_distance,x_span=FP_bragg_grating_period*FP_bragg_grating_num+margin_distance*2,
                      y_bottom=y1-FP_width[i]*0.5-margin_distance,y_span=FP_width[i]+margin_distance*2)
            rect_disp(x_left=x1+FP_length[j]/2+margin_distance,x_span=-FP_bragg_grating_period*FP_bragg_grating_num-margin_distance*2,
                      y_bottom=y1-FP_width[i]*0.5-margin_distance,y_span=FP_width[i]+margin_distance*2)
            
        elif layer_name == 'topcontact-win':
            rect_centre_disp(xc=x1,yc=y1,x_half_span=FP_length[j]*0.5-FP_contactwin_edge_dist,y_half_span= top_contact_width*0.5)
        elif layer_name == 'topcontact-region-drc':
            rect_centre_disp(xc=x1,yc=y1,x_half_span=FP_length[j]*0.5-FP_contactwin_edge_dist+4.8,y_half_span= top_contact_width*0.5+2.5)
        elif layer_name == 'bottomcontact-win-region':
            rect_centre_disp(xc=x1,yc=y1,
                             x_half_span=FP_length[j]*0.5+FP_bottom_contact_window_reflector_dist+FP_bottom_contact_window_width,
                             y_half_span= FP_width[i]*0.5+FP_bottom_contact_window_dist+FP_bottom_contact_window_width)
        elif layer_name == 'bottomcontact-win-diff':
            rect_centre_disp(xc=x1,yc=y1,
                             x_half_span=FP_length[j]*0.5+FP_bottom_contact_window_reflector_dist,
                             y_half_span= FP_width[i]*0.5+FP_bottom_contact_window_dist)
            rect_disp(x_left=x1-FP_length[j]/8-FP_contact_drc_dist,x_span=FP_length[j]/4+2*FP_contact_drc_dist,y_bottom=y1,y_span=1275-y1)
        elif layer_name=='metal-topcontact':
            rect_centre_disp(xc=x1,yc=y1,x_half_span=FP_length[j]*0.5-FP_contactwin_edge_dist,y_half_span= top_contact_width*0.5)
            rect_disp(x_left=x1-FP_length[j]/8,x_span=FP_length[j]/4,y_bottom=y1,y_span=1275-y1)
            rect_disp(x_left=x1-FP_top_pad_width/2,x_span= FP_top_pad_width+50,y_bottom= mesa_top_y+50,y_span=1275-y1)
        elif layer_name=='metal-bottomcontact-region':
            rect_centre_disp(xc=x1,yc=y1,
                             x_half_span=FP_length[j]*0.5+FP_bottom_contact_window_reflector_dist+FP_bottom_contact_window_width+5,
                             y_half_span= FP_width[i]*0.5+FP_bottom_contact_window_dist+FP_bottom_contact_window_width+5)

```
